# Max Absolute scaling 

```python
dataset = {
    'Age': [35, 45, 28, 50, 40, 33, 55],
    'Income (USD)': [50000, 75000, 40000, 90000, 60000, 48000, 85000],
    'Number of Children': [2, 1, 0, 3, 2, 1, 2]
}
```

```python
dataset = pd.DataFrame(dataset)
dataset
```

In max absolute scaling we divide the value by the max value to get value to occur in between Zero and One

```python
dataset['Income (USD)']
```

this data start from 50000 and goes to 85000 if we normalize it this data will be more simpler for our models to train with

```python
dataset['Income (USD)'].describe()
```

```python
sns.distplot(dataset['Income (USD)'])
```

```python
income = dataset['Income (USD)']
dataset['Income (USD) Scaled'] = income / income.abs().max()
dataset['Income (USD) Scaled']
```

```python
sns.distplot(dataset['Income (USD) Scaled'])
```

As we can see that the chart remained the same for condition that it will remain in range form 1 to  0
