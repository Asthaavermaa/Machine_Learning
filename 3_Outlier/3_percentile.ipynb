```python
# Importing suitable libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
```

```python
# Initializing the dataset
data = pd.read_csv(
    "/home/amandeep/Documents/DataAnalysis/dataset/archive/winequality.csv"
)
```

# Percentile method -

### Trimming

### Capping


## We can see that our data will have to in range of the 99% of the data and the 1% of the data


### Now we are gonna set the upper limit and lower limit

```python
sns.boxplot(data['volatile acidity'])
```

Now that we can seen the outliers we have to treat them

```python
# Setting the upperlimit
upperLimit = data['volatile acidity'].quantile(0.99)
lowerLimit = data['volatile acidity'].quantile(0.01)
```

```python
upperLimit , lowerLimit
```

# Trimming

```python
# Find the outliers 
newData = data.loc[(data['volatile acidity'] < upperLimit) & (data['volatile acidity'] > lowerLimit)]
print("Old data: ", len(data))
print("New data: ", len(newData))
print("Outliers are: ", len(data) - len(newData))
```

```python
plt.figure(figsize=(14,7))

plt.subplot(1, 2, 1)
sns.boxplot(data['volatile acidity'])

plt.subplot(1, 2, 2)
sns.boxplot(newData['volatile acidity'])
```

# Capping

```python
newDataOne = data.copy()
```

```python
newDataOne.loc[newDataOne['volatile acidity'] > upperLimit, 'volatile acidity'] = upperLimit
newDataOne.loc[newDataOne['volatile acidity'] < lowerLimit, 'volatile acidity'] = lowerLimit
```

```python
plt.figure(figsize=(14,7))

plt.subplot(1, 2, 1)
sns.boxplot(data['volatile acidity'])

plt.subplot(1, 2, 2)
sns.boxplot(newDataOne['volatile acidity'])
```
