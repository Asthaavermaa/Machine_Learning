```python
# Importing suitable libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import datetime

warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('/home/amandeep/Documents/DataAnalysis/dataset/archive/Traffic data.csv', nrows=100)
```

```python
data = data.loc[:, ['Datetime']]
```

```python
data
```

```python
data['Datetime'] = pd.to_datetime(data.Datetime)
data.head()
```

```python
data['year'] = data['Datetime'].dt.year
data['month'] = data['Datetime'].dt.month
data['day'] = data['Datetime'].dt.day
data['quarter'] = data['Datetime'].dt.quarter
data['datOfWeek'] = data['Datetime'].dt.dayofweek
data['week'] = data['Datetime'].dt.week
data['isWeekend'] = np.where(data['datOfWeek'].isin([5,6]),1, 0)
```

```python
data.head()
```

```python
data.sample(frac=1).head(10)
```

### Extract Time Features 

```python
data['hour'] = data['Datetime'].dt.hour
data['minute'] = data['Datetime'].dt.minute
data['second'] = data['Datetime'].dt.second

data.head(10)
```

### Extract date and time

```python
data['time'] = data['Datetime'].dt.time
data['date'] = data['Datetime'].dt.date
```

```python
data['difference'] = datetime.datetime.today() - data['Datetime']
```
