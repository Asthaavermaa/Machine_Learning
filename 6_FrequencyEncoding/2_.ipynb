```python
class BinaryEncoder:
    def __init__(self):
        self.encoding_map = {}

    def fit(self, data):
        unique_values = sorted(set(data))
        num_bits = len(bin(len(unique_values) - 1)[2:])  # Calculate number of bits needed
        self.encoding_map = {value: format(index, f'0{num_bits}b') for index, value in enumerate(unique_values)}

    def transform(self, data):
        encoded_data = []
        for value in data:
            encoded_data.append(self.encoding_map[value])
        return encoded_data
```

```python
# Example usage:
encoder = BinaryEncoder()
encoder.fit(data.CITY)
encoded_data = encoder.transform(data.CITY)
```

```python
data['Encoded_data'] = encoded_data
```
