```python
# Importing suitable libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv(
    "/home/amandeep/Documents/DataAnalysis/dataset/archive/winequality.csv"
)
```

## Importing sklearn standard scaler

```python
from sklearn.preprocessing import StandardScaler
```

## Initailizing  standardscaler

```python
scaler = StandardScaler()
```

```python
scaler.fit(data[['alcohol']])
```

```python
scalerData = scaler.transform(data[['alcohol']])
```

```python
scalerData = scalerData.reshape(-1)
```

```python
plt.figure(figsize=(14, 7))

plt.subplot(2, 1, 1)
sns.distplot(data['alcohol'])

plt.subplot(2, 1, 2)
sns.distplot(scalerData)
```
