```python
plt.figure(figsize=(14, 7))

plt.subplot(2, 1, 1)
sns.distplot(data["volatile acidity"])

plt.subplot(2, 1, 2)
sns.distplot(data["alcohol"])
```

# For scaling the data we need to apply the formula

# ZScore Method

# {Value - Mean()} / std()


```python
data["alcohol scaled"] = (data["alcohol"] - data["alcohol"].mean()) / data[
    "alcohol"
].std()

data['volatile acidity scaled'] = (data["volatile acidity"] - data["volatile acidity"].mean()) / data['volatile acidity'].std()
```

### Now we will plot our graph for our scaled values

```python
plt.figure(figsize=(14, 7))

plt.subplot(2, 1, 1)
sns.distplot(data["volatile acidity scaled"])

plt.subplot(2, 1, 2)
sns.distplot(data["alcohol scaled"])
```
