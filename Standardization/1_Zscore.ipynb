```python
# Importing suitable libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv(
    "/home/amandeep/Documents/DataAnalysis/dataset/archive/winequality.csv"
)
```

```python
data.head()
```

```python
data.describe()
```

```python
sns.distplot(data["fixed acidity"])
```

```python
scaledData = data.copy()
```

```python
for cols in ["fixed acidity", "pH"]:
    scaledData[cols] = (scaledData[cols] - scaledData[cols].mean()) / (
        scaledData[cols].std()
    )
```

```python
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
sns.distplot(scaledData['fixed acidity'])

plt.subplot(1, 2, 2)
sns.distplot(scaledData['pH'])
```
